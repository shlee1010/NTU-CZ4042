```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {
    0.001: {0.3: 0.6571, 0.4: 0.6658, 0.5: 0.6781},
    0.0005: {0.3: 0.6697, 0.4: 0.6700, 0.5: 0.6646},
    0.0001: {0.3: 0.6628, 0.4: 0.6663, 0.5: 0.6690}
}

dropouts = [0.3, 0.4, 0.5]
learning_rates = [0.001, 0.0005, 0.0001]

fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('CNN Cross-Validation Results by Learning Rate', fontsize=14, fontweight='bold')

for idx, lr in enumerate(learning_rates):
    ax = axes[idx]
    accuracies = [data[lr][d] for d in dropouts]

    ax.plot(dropouts, accuracies, marker='o', linewidth=2, markersize=8, color='#ff7f0e')
    ax.set_title(f'Learning Rate = {lr}', fontsize=12)
    ax.set_xlabel('Dropout Rate', fontsize=11)
    ax.set_ylabel('Average Validation Accuracy', fontsize=11)
    ax.set_xticks(dropouts)
    ax.set_ylim(0.64, 0.69)
    ax.grid(True, alpha=0.3)

    for d, acc in zip(dropouts, accuracies):
        ax.text(d, acc + 0.002, f'{acc:.4f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('cnn_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
