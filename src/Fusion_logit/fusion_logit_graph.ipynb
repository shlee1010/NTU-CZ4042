```python
import matplotlib.pyplot as plt

data = {
    16: {1: 0.7055, 2: 0.6984, 3: 0.6959},
    32: {1: 0.7051, 2: 0.7063, 3: 0.7027},
    64: {1: 0.7048, 2: 0.7052, 3: 0.7058}
}

widths = [16, 32, 64]
depths = [1, 2, 3]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Logit-Level Fusion Grid Search Results (LR=0.001)', fontsize=14, fontweight='bold')

ax1 = axes[0]
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']
for idx, depth in enumerate(depths):
    accuracies = [data[w][depth] for w in widths]
    ax1.plot(widths, accuracies, marker='o', linewidth=2, markersize=8,
             label=f'Depth={depth}', color=colors[idx])
    for w, acc in zip(widths, accuracies):
        ax1.text(w, acc + 0.0003, f'{acc:.4f}', ha='center', fontsize=8)

ax1.set_title('Accuracy by Hidden Width', fontsize=12)
ax1.set_xlabel('Hidden Width', fontsize=11)
ax1.set_ylabel('Mean CV Accuracy', fontsize=11)
ax1.set_xticks(widths)

ax1.set_ylim(0.694, 0.707)
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2 = axes[1]
colors2 = ['#d62728', '#9467bd', '#8c564b']
for idx, width in enumerate(widths):
    accuracies = [data[width][d] for d in depths]
    ax2.plot(depths, accuracies, marker='s', linewidth=2, markersize=8,
             label=f'Width={width}', color=colors2[idx])
    for d, acc in zip(depths, accuracies):
        ax2.text(d, acc + 0.0003, f'{acc:.4f}', ha='center', fontsize=8)

ax2.set_title('Accuracy by Hidden Depth', fontsize=12)
ax2.set_xlabel('Hidden Depth', fontsize=11)
ax2.set_ylabel('Mean CV Accuracy', fontsize=11)
ax2.set_xticks(depths)

ax2.set_ylim(0.694, 0.707)
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('logit_fusion_results.png', dpi=300, bbox_inches='tight')
print("Image saved: logit_fusion_results.png")
plt.show()
```
