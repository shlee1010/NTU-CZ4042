```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = '/content/mlp_cv_results.csv'
df = pd.read_csv(file_path)

df['hidden_depth'] = df['hidden_depth'].astype(str)

g = sns.relplot(
    data=df,
    x='hidden_width',
    y='avg_accuracy',
    hue='hidden_depth',
    hue_order=['2', '3', '4'],
    style='hidden_depth',
    col='learning_rate',
    kind='line',
    markers=True,
    height=4,
    aspect=1.2,
    legend='full'
)

g.set(xticks=[64, 128, 256])

g.fig.suptitle('MLP Cross-Validation Results by Learning Rate', y=1.05)
g.set_axis_labels('Hidden Width', 'Average Validation Accuracy')
g.set_titles("Learning Rate = {col_name}")

g.legend.set_title("Hidden Depth")

output_image_file = 'cv_results_plot_ordered.png'
plt.savefig(output_image_file, bbox_inches='tight')

print(f"\nPlot saved to {output_image_file}")
```
