```python
import pandas as pd
df = pd.read_csv("/Users/sanghyun/Downloads/fma_metadata/tracks.csv")
genre_map = pd.read_csv('/Users/sanghyun/Downloads/genre_mapping.csv', header=None)
```

```python
genre_map.columns = ['genre_key', 'genre_ids']
genre_map
```

```python
df = df.rename(columns={
    'Unnamed: 0': 'track_id',
    'track.8': 'genres'
})
```

```python
df = df[['track_id', 'genres']]
df = df.iloc[2:].reset_index(drop=True)

import re
df = df[df['genres'].astype(str).str.match(r'\[\d+\]')]

df
```

```python
print(df.dtypes)
print(genre_map.dtypes)
```

```python
import ast

def to_list(x):
    if isinstance(x, list):
        return x
    if isinstance(x, str) and x.startswith("[") and x.endswith("]"):
        return ast.literal_eval(x)
    return [x]  # fallback

df['genres'] = df['genres'].apply(to_list)
genre_map['genre_ids'] = genre_map['genre_ids'].apply(to_list)
```

```python
valid_roots = {10, 12, 1235, 13, 14, 15, 17, 2, 20, 21, 3, 38, 4, 5, 8, 9}
mask = ~df['genres'].apply(lambda genres: all(g in valid_roots for g in genres))
```

```python
def genre_id_to_root(genre_id, genre_map):
    for _, row in genre_map.iterrows():
        if genre_id in row['genre_ids'] or genre_id == row['genre_key']:
            return row['genre_key']
    return genre_id
```

```python
def fix_genres(genres, genre_map):
    return sorted(set(genre_id_to_root(g, genre_map) for g in genres))

df.loc[mask, 'genres'] = df.loc[mask, 'genres'].apply(lambda x: fix_genres(x, genre_map))
```

```python
# Print count of unmapped genres
unmapped = df[~df['genres'].apply(lambda genres: all(g in valid_roots for g in genres))]
print(unmapped)
```
